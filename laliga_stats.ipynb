# 2) Print all the teams which have started playing between 1930-1980.

```python
df['new_Debut']=df['Debut'].astype(str).str[0:4]
liga=df.astype({'new_Debut':'int64'})
liga2=liga[(liga.new_Debut>1930) & (liga.new_Debut<1980)]
liga2[['Team','Debut']]
```

# 3) Print the list of the teams which came top 5 in terms of points.

```python
laliga=df.astype({'Points': 'int64'})
top=laliga.sort_values('Points', ascending=False).head(5)
top[['Pos','Team','Points']]
```

# 4) Write a function with name "Goal_diff_count" which should return all tems with their Goal Difference with same function.

```python
df['GoalsFor']=df['GoalsFor'].astype(int)
df['GoalsAgainst']=df['GoalsAgainst'].astype(int)
def Goal_diff_count():
    df['Goal_diff_count']=df['GoalsFor']-df['GoalsAgainst']
    return df[['Team','Goal_diff_count']]
Goal=Goal_diff_count()
Goal.sort_values(by='Goal_diff_count',ascending=False)
```

## Find the team which has maximum goal difference.

```python
Goal_diff_count().head(1)
```

## Find the team which has minimum goal difference.

```python
Goal_diff_count().tail(1)
```

# 5) Create a new column with name "Winning Percent" and append it to data set. If there any numerical error, replace it with 0%.

```python
df['GamesWon']=df['GamesWon'].astype(int)
df['GamesPlayed']=df['GamesPlayed'].astype(int)
df['Winning Percent']=(df['GamesWon']/df['GamesPlayed'])*100
df['Winning Percent'].fillna(0,inplace=True)
df[['Team','Winning Percent']]
```

## Print top 5 teams which has the highest Winning Percentage.

```python
df[['Team','Winning Percent']].head(5)
```
